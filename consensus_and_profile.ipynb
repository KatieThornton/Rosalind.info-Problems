```python
sample_input = """>Rosalind_0337
TCGGGCACCGAAGGTGCGGTTGGACACATACCAGACGTCGGTCGGAATCAAGCCATCCCC
GCGTAAACCTTGCACGTCCTAACACCATCCGTACGACTATTCTCATCTAGGGAATCTCGA
AATATCTAATGCGACTTGCTGCATCGGATGCCAGTAGCATCTTAGTTTAACTGTAACCCA
CAACTCAGGTACGACATAGGGGACTACGCACTCTTTCATCTTTGGCAGAGAAATGACGGG
GCCCTTATAGGTTGAGAGTTCTAATGATATCCCTCCAGTATTAGACTTACGGTCGGACAG
CCGAGTCTTGAAATCAGATAAAGGGTGTAGAGTACTGCTGAGTTTTTTATATGGACTCAA
ATACAGAGACTAATGGTCTTTATCCGTCCCGAGACAGTACAAGACACTGCAAGCTGCTGA
GCGTATAAGATTGCAGACAAGGCGCTTACTTCGGCTCTACGCTTAGCAGGGTATCAGGTG
GACTTCTGAGTGGTTCGTTCGTCCAAAAAGCAGATCAGAGGGGACGCGCTGGGGCCTACG
CCCGCTTTGCAGCATAATAGAGGTCTGTAATGCTAAATCGGAACATCACTCGTTCGTGGT
AGCCACCCTCCCTGCCGCGTCCACTAATCAAGAGGACAGTCGGTATAGGGGCTAATCGTT
TGGCTTATGCGGAGCACTCCATCGCGTTGGGATTTAGTATGGGATCCAAGCTGCGGTTGC
GACCCACCTAGGAGACAGTCGAGTCCTGGCTGCGGGTGAATTGAGGATCCTACCCATGTC
GGTTACGGGATAATCTTATCACAGCATTCTGAGCTCCGCTGTCACAAGCTTATACACGAG
ACCCTTAATTTGTGAAAAGATGTTCCTTGGAGCACTTGTAGTCGTGCATGGTCTTTAATG
ATAG
>Rosalind_1782
AGTAGCCAAAAAAAGACCTAGAATGCTCTTAACTCAGTCGATGTCGTTTGAGTGGTTCAA
GCTGCAGGTTTACGCGTGGACGGGGGACACGGCAAAGTACAATGGTGCACATGGCGTCGT
CCGTCTTCCTAGCTTGCGAGCGCAGCATCTCCCTCTTTGGTCATACGCGAGTTTTTGTGA
CTTCGGCGCGGAGAGTTAGGCGTCGTTATCGTAGTAGGTTGGCCTGATGCATAACGCCAT
CCATGTGGATACAGTACCTCCGAGCAACTGTACGATGGAGGGCCGGTTCATTACCCAGAG
TACCGCCGGCGATGGAAGCGACAAAAGGAGCCCCCAAAGTGGTGCGAGAAGCATCTAACC
TAATGTTTATCGTGAACTTGCGCATGACTGCATGAAATGTTAAATGACAGCAACGAGGTC
CCCCCTGAGAGAAGTGAATTCATTAGTCGGATAATGACACTTTTAGGGGGACGCCGTAAT
ATAATTAGCCCTCCGCTTACCTATTTGAGCCGTCTTGAAACCAGAGACAGCCCTATGATT
TGCTCTGGGCGGACCGTACAGTGATAGCATTTAGTGAAGCGACACATTTAAGTACTAGAA
AGATTAAAGCGAGTTCTTCAGGTGGCGGGTCTTAGTCCGACAGGTTGCAGCCGGGCGCCT
CAGTCTAGGTGGCTCAGTGATACGCCAGGCAGGCTGAAAATGTTGCAAACCGGTTAAAAC
CTCTACTTGAGCTGACCCGGTTCGGTTCGTTAACCATCACGATAGAAACACGAAAGTGAG
TGATATGAGCATCATATGTTAGTTGAAAGAACGGAAGGGCGGCGCGACAATCTGCCCGCT
GTCAGTAACATATTCGCGTGGCCAACCGTATAGATCTCCTCGGATATGAGATGCCGAGTT
AGTA
>Rosalind_8974
GGCGGGCTTCAACCATGAGGATAATTCGACCACAGCCGGGTCCCTTCAGGAGAGTATCAG
AGAAGGTCGCGTTTCGTTATTTTCCATTGCGCGGTCCAATGGACGCACACAGTATAGCTG
TCCATCCTACGCTGCTCTGAGCATGTCCTTAATCCGAAGAGGTGCTACTAAGTAATACTC
ATTAAATCATGTCCTAGGGGTCTCACATATACCGCCAGAATGATAAAGACGTGTTCGAGC
ATATAAAATAAGACTTGCCAGGACTTAATGTACCGGATCACCCTCGTCGGTTGATAGTAC
TGCTGAGTGTCTCTCGTGAACTGCTATCAAGGTGAGCAGGAAAAAACTGTTCGGTATCAT
GAGGGACATATCTAGCGTGCCTTCACACTAGCGAACCTGTATTTGACCCTAGTGATTGAG
CGTGCCGGATTGCGCGCGCACCAGTTGCCCTCACAACGCAGAGCTGCTTGAGCGACTGGG
ACAACCTGTCACCTCGTCGCCGCCTACGCCTTAAGTGTCGCGGGATTTTGCGGGAACTGG
ACATTATACCGCTGTACATAGCCACGCGCCCTGTTAACTCGCTGTACCCGATTCGGCCGG
GACCCCCAAAGACAGGCTCCTCCGAATTTCCGCTCTTAAGGCGGTTAGGGCGCAACAAGT
TTGTGACAAGTCGGCAAATTCATTCCCAAATTGCATTATGCCCTCACGCTCGACCACAGG
TCACGTATAACAGGGTCGAGACCGCTCATGTATCGGCCTACGATCCAAGTACGTTAGAAT
TCATCATCTCGCGTGAATCCAATTTCGGAATGCTGCTCGGGGAGAATCATGTCCAACGCA
CATATTGCACAGGAGGGCGCTGGAATGTTATAGAGCATAGGGTAAACTCTCAGAGACGTC
AAGG
>Rosalind_4070
CTAGTAATTATTAACTTTAACCATCTGATACATTGAATTCAGCATAACGCTACGTGCAGT
CGGCCTAGACGTTGTCTTTAGCACCGGGCAAGGGACCGAGCCTAGCCAGTGGATCCAAAG
GGGTCTTGTAGAGTATACCGCCCTCAATCGCAGGAAGGGCCATTTAAGAGGGTTTGCCAG
CGAACCCGCGAACCCTAGCATGACTTAGACCAAGGTTATACCAGAAAACCACCTTAGGGG
TACCCGGAAAGTCCAATTATTGTATAATACCTGATCGGTTGTACTGGCGTCGAAAGACTC
ATCGAGCAATTTATACCAAAGTGAGTTAAAATGCGTTGGAAGACTATATTCAGAAACTAT
GCGGTTGGAAAACTTGGACCCTACAATACTGGCCAATAGATGGCCTACGTTGCAAAATGA
GAAGGCTGGGGTGCCGGGGGCTGGAGCTCAACTGCCAAGCAACCCCATTACGGGAGGATG
AATTAGCCATTTGGCACGTCCGGCGTGGCTACGTCGGCTCAGCTATGTGTCCCTTGTTGA
GTCTTCTTCATCCAGAGTCCGCCGTCTCGCGGCCGCGGTATTTACGTATCGTTAAGGATA
TAGTGGCACTTACTCATCCTTTGGCGTCATACGACGCAGGCCGCGTATCATTATCACGTT
TGCGCAAAGGTTCATTCACAGATATCTACGACTGTTAACAAAACGGCGCCACATACTCAA
TTGCCTTTAATCGGCCTGGAGAACTGGACTGCATGCGACGGTTCGCGGGACGGCAACTAT
CCCTTGTTTTTGGTGGGAGCCCAAGTTCCTTTTATCACCCTGAATGCCAGATGTTTTGGA
CGCAACCGGTTCTCTAAACGGAACTCCCCCTTCGTTTAGTAGGCTAACGATTTTACGTGC
GACC
>Rosalind_3152
CAAACAGCAGGGTATAGTTGCGTTTGGAGTCGATGCATCCAACTAGTGTACGCCTCACTT
TGTTCCCTGTTGGATAAAATCGTGCTTTGGTCAATCATAACGGCAGCTGATAAGAATGAT
CAACTAACCCTGTACTCATTGGTTAGAGCACGTCCAAAACACAAACTGTTGGAAACCTCC
GGCGGTTATGCGCCAACATACGCTGAGCATGTTACGGCGCTTGTTTCAGGGTTAGCTCCG
CGACCTACCAACCTACCCAACCCCGCTCTGGAGGCATAATAGTATCCTCTAGTTCGGGTA
ATGCGCTCCGACTGTAAGGTTTATAGTTCCATTCGTTAGCAGGCTGGGGTCTCACGACAC
GAACATATACTCGGTGAGCGCTAAGAGGACCCCCCATATGGAACCGAGCAAGTTTGAATT
AAGAGGTCCTGAGGTGTGGCTATCTAGATACTGATCTGAGGCTAGGCAACAAGGAGGCCA
AGCGATCTCCGAGCAGCTGGCTCATTAGGTAGGGCATGTGCCAGGTCTTAAAACGTATAC
CTCTGGCGATCTACCCTTATGGACTCCCGAAAGCTCGGTGTACTCACGGTTAGATGTCAC
CTATGCTTCAGACGTCCAAGTCGCATAACGAACTTGGGTCATTGGTCATTAGAATGAGTC
GCACCTGCCGGTACGCGCTAGGGGACTTTCATTTCTAGTATTTATGAGAGCCGAAAACGG
TAGTAAACCTCTCACTTGCTAGTTTCTTACCTGCAGCTCACCATCGAGCACGCCTCTGAA
CGTTCTCGCCATGCGCAGCGTCACACTACGTCGTATCCAAGTCAGCCGAGTGAGGCGGTA
TCAACGAAGATCCGCAAAGAACAGCATTAATTCATTTGATAACATTCTACAGGGTAAGGG
CAGG
>Rosalind_1475
ATGCGCGACTTAAGCGGGGAAACACAATCTACTGGGTGATAAACTGCCTTACGTGACTAG
TCAGGACACATAACCTAATATCTTCTATTCTGACCCGGGGAGGATAAACTCCTATACTGC
TGGAAGTCAACTCTACACGAAAACTAAGCGCGAAGCGGGCGAATCGTATCAAGACTGACT
ACACGAGAACAACGATCGTTCTGCAGGAAGGACCTAAGTATGATAGTGATTTTTTCATAA
AGGGGGACTCTGAATAAACCTATGCTTGCACCAGCGTATCTAAGACATTTTTCCCCCAAA
CTCTTTAATTTAACATGCTTAGTTGCTCCAAACCATGACTTACACAAGCGACCCCGACTA
TTCTTGAACTAAAGTCAAGACCCGGTCGATGACTACAGACGATTCTTGAACAGCAGGATC
TACGATAAAAGCTAATGACCATCAAATCACCTCAGCCGAAGTCCACTCCTTCCCGCATAG
ATAGCCTGCCCCTGATTGACTTGATCAGATTCGGGCATCCTTCAATTGATTTCTATACCC
GCAAGTGGAGTCACGCGTCGGACAACAACATAATTGCCGTAGAAAGTACCCAGTGGGAAC
GAAGCTATACCATTTTACAGAGTGGTTATAACACGATCCGCATGCAGGGCAGAGGTCGTA
TTTATGGTACGTCGGTAGCTATTGTACATCGAACCAGCTAGGCATGTTATTTGGTGCTTA
CCACCGCTTGTGCGGAAGTGGTTCACCTCCAATAAAAACTGAGTAGCATTGTCTTAGACT
TGTGTTCTGCATATTCACTACGGCTGCTAAGCTTAGATGGTGTGCCGGGCGGGTAGCTAG
GTTAGAATGTCACCGGGGACCACAAACTGTGGGGGTAGCCACAAGAGCGTCAATCCTAGG
TTCG
>Rosalind_7767
TGATTAACCGCTAGCGTTGAACCAAGATTCAAACCTCCTTGAATCACGCTGGGACCGGGA
CATCGGCAGCATTCCCTGGTACGTCACCAGCTGGAGATGATCTGGGCTAAAGACCTTGGG
GGTTACTTCAAACGACCTGGATACTGTCGGCAGGTGCTCTGACAGGCTACGGCTACGAAC
GCTTGTGGCGTCTCTTTTTTAAGCATTAGACACTATGAACTGACCTCGTTAGAGCAGGCA
ACTCACTGTGTACACGCCTTGTTTCATTCATAGCTAGAGTCCGGGAAGATGTCTGTATGA
CTAAAGCAAGCGTCTAGTGAGGAGCGACCTGACCGACAATGCCTATGTTCCAAAGACTGG
GCTGAATGGTTTTTCAAAAGGTTGAGAAAATCTGGCGGTAGGTATATTCTAACGTTTGTA
ATCGGTGAATTCCCCCATTTAGTTATACCAATCACTGCGCCTACAGCATTAGTGTAAGCG
TCGTTAATGTAGAAGCGAAGTTTCACGCTTTTGAGACCTGAATCAGGGACTAGCGGACTG
TACGCTTACTTTGACAAACGTGAGGCACCTAGCTCCCATGGGAAAACGCTTCCACGATGA
TGAGACAGAGTTAAGGCTGTGCAACCCCGGGGTTTGATTGGCCCTACATTCGCCATTAGC
CAGGACCTGTTCCTGATTAAGACTCCAGTCGTCTCGGCACTTCTACTTTTCTCCCAGCAA
CCCTTTGTCTGAGTAGAACATGCCACGCGCGCAATGCTGGACCAAAACCCACCTGAAGGC
CAGAACAGATATATCTAGAATCTTATGAGCGCCCCACCACCACAAGTCTTCTGCGGTGCT
TTGTTCTCGCCTCCGAACGCATCAATTGCGTTTTTCACTACTCCGAGAATCCTGCGGCAA
GAGG
>Rosalind_5295
ATATGAAGAACACAAATGTGGCTAAGTCGTCCCCATATGTTCCGAGGTGAACAGACTCTC
GTAGGTATAGATAGTTTTTAATTCTATGTTCAAGGCCGAGGACAGTATCGCAAAGTTGCG
AATAGCCCTTATCCGGGTAATGCAACATTCGCGCTCTGGCTAATGGGCGGCCGCCCGTTA
TGCACGTCAGAGCCCCGAAGTGTCGAGTGCTTGAATCTTATTTGTTGCTAGATTAGAAAC
CCTTCCTACGGTTTTCTCCGCTCCACGACATCTAGCCCCTACCGCAGCAGACGGGGCTTG
AATCGATACGCAAGGTTGAAGACGTACTTCTACAGGTTCGAATGAAGGTTTTAGAGATCA
TGACGCCCGTAACTTGGACCTCACTGGAGTAATGCAAGTGAAAATGGACTCTCTCCCGGA
CAACGCGCCGAAAGGTAGGCGAAAAACTGCATAATGAGTGACGGATAGCCGCCGAGAGGG
GTTTCTATGCGTGGCGCCCAGCAAAACAAGGATGGGCCCAGCCGTCGAGATGCAGTGTCA
CATACAGTACCCAGCGGTCAACATCGGACAGAGCTGACGACACCGCGCATACACTTTGTT
TTTGAGAGGCTCTATGTCCCGAATCACAAGCGATTGTACTTTGCAACCGGTGGTGGAGTA
AACGACGAGTCAGTAGCTTTAACGCCACTTTCGTAGGGGACGAAACAATGCATACTGATA
GGAGTTGCTGAAGCGCATTTGAAGGGCACGCTCCCTGCACATTTCCGCTCTATTTTAGCA
ATCTTGAGACTCAAAACCCACCTGTAGATATCATGGCCTCGCACTTGAACCACGGCTCAT
CGCGACCCCCGCTCCCTGCGTGAATCCGGAGACCCTTAATCACAGGGGTGTGAGATCGGA
GACA
>Rosalind_3071
ATACAGCGAAAATCTAAGTTCTGAAATCGGTCGCCCTTATGGGCTTCTCGGGTATAACTC
TCCGCGACTGTTCTTACTCAAACAACAGTTCCGGATACAGTATAGACCCCGTGATAAATA
ATTTCGGACGGTCGTACGCCGTTTAACCACGGCCCCTGCAACTGGTGACTTACGTGTCAA
TGCGAAGGTTTTCCCCGGCTAGGTTAAATTGTTTCTGTGCGAGCCCGTTTGTCGCAGGAA
GAGAAGTCTAGAGCATCGCAGACCGACATAGGGCCGGCCGATTACGAACTGGCGGAGTAG
TAGTAACATCCCGTGCGGTCACGTAGACCATAGGACGTCTACAGGGGCTAGGTACAGGAT
CGAGGTGGGCGGCTTGCTGAGGACCAGGCCGGCCGGTAACTGATGGTCCCACCGTTCTCC
CCGCCCCTAACCCCCGTATCAATAGTTGCTGGCAATCGGATGCACCCTCATGGCCTCCAA
GTCGTAGTCGTGGACCCATGGTGTTGATTCAAGCTTTTTATTCCCCATAGTGCGATCGCG
TCCGACGGGGGCTGGGCTATTCATAGCCGGGAATCCGAGACAGTGATTTATTCGCAGAAG
TATTTCTGACCCGATCTCAGGCTCACAACTGCTGGTCAAGATGCTATACAAGCATTACAC
AGAGTAATCTCAGTACGTCTGTACGGTTCAGATTCTGCGAGCATGGTCTGTCACGGGCAA
CCTGTCTAGCCTAGCATGCACGTGTTTAGTTAGCGTTGCAACCATAAAGCAGATAACAAT
AGACATCAATACAGAAGCCTGATGTGCAAACCCGAGAACCCCAGCGATGGCCGGCTAAAG
CCCTTTCTCGACGCCCGAGTTGCGAAACCACGATCGCTTGATGCACCTTTTCTGCGCATA
ACAG
>Rosalind_3428
GCCATGGTAAGGCTAATGTTTCATACTACGACACAGCTCCGGACCTCTTATCCCGAGCAG
CTGTTCGAAACACCCGGATCGCATGCGTCGTCCGATACTTTCTATCTTCGCCAGGGTATC
TCCCGGACACATTCTCGTTTATTTAACCGCGTTGCTTTGTATGCAGCAGATATTAGATGG
ACGTGACAACCGCTACATTTTCAGAGTATAATTCCCAAAGACCAGGCGTATTGCAGCTTT
ACATGTATTGGTGCAAATCTCCTCCGCTCGTACAACACCGCATTCGTGAGAAATGCTCCC
CACCTACCCTCGCTCAACTCTGCTACAGCATCAATCAGTTCATGAAGCCTTACACTGCAT
AATCTTGCTCCTTGTAAACCAGGATTGATATTAAACGCCTGAGGCTTAGGGACCTAGGTA
CACAACGAGAGACTGTGCATGCCGGGCGTCGATTTGCTTGACCGTGCTGCCTCGGTTGTT
AATACCTATACTACTTAAAACGGTCGTGACCACTTCAGGATTCGGGTGATTCGTCGCACA
CGGACGTAGATTACGTCTATCTCCCCAGGGTGGGTAGCACAGATGCGGAACACTGTGTTA
AGATAAGCTATTAATGGGCTCGCTAGAGTAGCATGAAATTGGTCATTTGTCGCTCCGACC
TGTTTATATTTGATAGCAATTCAGGGACCGAGCCTCAGATTTGAATTAAGCTGCCAGGCT
TTGATCAACCCAGGAGTTTGTTATACCGAGAGATGGTTTTTACCTCGTCGTTGTGCGGTA
CAAAATAATTTTCTTGGTCTACCGCGAAGCCATCTCATTCCCAGCCGCACGTCACTATTT
CGCTGACCTGATCGGAAGGAAACAGGCCTTCCACATACGACAACACGGAATGCGGATAAC
CTTC"""


def consensus_string(fasta):
    dna_strings = []
    current_dna = []

    for raw_line in fasta.strip().splitlines():
        line = raw_line.strip()
        if not line:
            continue
        if line.startswith(">"):
            if current_dna:
                dna_strings.append("".join(current_dna))
                current_dna = []
            continue
        current_dna.append(line)

    if current_dna:
        dna_strings.append("".join(current_dna))

    if not dna_strings:
        raise ValueError("No DNA strings found in FASTA input.")

    L = len(dna_strings[0])

    matrix = {
        'A' : [0] * L,
        'C' : [0] * L,
        'G' : [0] * L,
        'T' : [0] * L
    }

    for dna in dna_strings:
        if len(dna) != L:
            raise ValueError("All DNA strings must have the same length.")
        for index, char in enumerate(dna):
            if char in matrix:
                matrix[char][index] += 1

    consensus = ""
    for i in range(L):
        consensus += max("ACGT", key=lambda base: matrix[base][i])

    print(consensus)
    for base in ['A', 'C', 'G', 'T']:
        print(f"{base}:" , *matrix[base])

consensus_string(sample_input)
```
